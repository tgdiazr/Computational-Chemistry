```python
from google.colab import drive
drive.mount('/content/drive')

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
import os # Import os module
from google.colab import files # Import the 'files' module from google.colab


# Specify the directory containing the data files
data_dir = '/content/drive/My Drive/Resultds-MD/New'

# Get a list of all .dat files in the directory
file_paths = sorted([os.path.join(data_dir, f) for f in os.listdir(data_dir) if f.endswith('.dat')])

# Plot settings
plt.figure(figsize=(8, 6))
plt.xlabel("Time (fs)", fontsize=20, fontname="Arial", labelpad=10)
plt.ylabel("Total Energy (eV)", fontsize=20, fontname="Arial", labelpad=10)
plt.xlim(0, 1100)  # Adjust x-axis limits as needed
plt.ylim(-50000, -66000) # Adjust y-axis limits as needed
plt.xticks(fontsize=20, fontname="Arial")
plt.yticks(fontsize=20, fontname="Arial")
plt.axhline(0, color="red", linewidth=1)
plt.axvline(0, color="black", linewidth=0.1)

# Loop through each file and plot the data
for file_path in file_paths:
    # Load data from file, skipping header lines and specifying delimiter
    # Use genfromtxt to handle invalid values, replacing them with NaN
    data = np.genfromtxt(file_path, skip_header=1, delimiter=' ', invalid_raise=False)

    # Remove rows with NaN values
    data = data[~np.isnan(data).any(axis=1)]

    # Extract columns (assuming first column = Time, second column = Energy)
    x = data[:, 0]
    y = data[:, 1]

    # Remove duplicate x values and corresponding y values
    x_unique, index = np.unique(x, return_index=True)
    y_unique = y[index]

    # Create a smooth curve using cubic splines
    x_smooth = np.linspace(x_unique.min(), x_unique.max(), 300)
    y_smooth = make_interp_spline(x_unique, y_unique, k=3)(x_smooth)

    # Plot the smooth curve with a label based on the file name
    plt.plot(x_smooth, y_smooth, linewidth=2, label=os.path.basename(file_path)[:-4]) # Label using file name without extension
    plt.fill_between(x_smooth, y_smooth, -100000, color="white", alpha=0.5)  # Adjust fill as needed

plt.legend(fontsize=15)
plt.grid(False, linestyle="--", alpha=0.0)

# Save the figure before showing or downloading
plt.savefig("plot.png", dpi=300, bbox_inches='tight')

# Download the saved figure
files.download("plot.png")

# Now you can show the plot in the notebook (optional)
plt.show()
```
